```python
def get_tags_from_url(input_url):
    # Get response from url
    response = requests.get(input_url, headers = headers)
    # Encode response for parsing
    html = response.text.encode('utf-8')
    # Soupify the response
    soup = BeautifulSoup(html, "lxml")
    # Get tags - they are contained in class "taLnk" as hyperlinked elements
    tag_elements = soup.findAll("a", {"class":"taLnk"})
    
    dict_of_tags = {}
    
    if (len(tag_elements) != 0):
        # Iterate over tag_elements and exctract those which have parentheses 
        # as they are the relevant items
        print("Yes")
        tags_list = []
        for i in tag_elements:
            strng = i.string
            if("(" in strng):
                tags_list.append(strng)
    
        # Create final list of tags
        for item in tags_list:
            items_list = item.split('(')
            name = items_list[0].rstrip()
            # Get value
            value = item[item.find("(") + 1:item.find(")")]
            dict_of_tags[name] = value
    else:
        # Second kind of web page with list of 'Things to Do' at the bottom of the page
        # Use selenium to automate browser functionality
        # Browser opens the input_url, scrolls till the bottom of the page,
        # clicks on 'Show All' button, the entire list becomes visible; then scrapes it
        print("Nope")
        #time.sleep(5)
        '''
        d.get(input_url)
        WebDriverWait(d, 20).until(EC.presence_of_element_located((By.XPATH, '//*[@id="component_3"]/div/div/div[12]/div[1]/div/div/div/div[1]/span')))
        d.execute_script("arguments[0].scrollIntoView();", 
                 d.find_element_by_xpath('//*[@id="component_3"]/div/div/div[12]/div[1]/div/div/div/div[1]/span'))
        d.execute_script("arguments[0].click();", 
                 d.find_element_by_xpath('//*[@id="component_3"]/div/div/div[12]/div[1]/div/div/div/div[1]/span'))
        html = d.page_source
        soup = BeautifulSoup(html, 'lxml')

        items = soup.findAll('a', {'class':'attractions-attraction-overview-main-Pill__pill--23S2Q'})
        '''
        time.sleep(5)
        d.get(input_url)
        show_more = wait.until(lambda d: d.find_element_by_css_selector("[class='ui_container'] div:nth-of-type(1) .caret-down"))
        d.execute_script("arguments[0].click();",show_more)
        soup = BeautifulSoup(d.page_source,"lxml")
        items = [item.text for item in soup.select("[class='ui_container'] div:nth-of-type(1) a[href^='/Attractions-']")]
        
        #tags_list = []
        #for item in items:
        #    tags_list.append(item.get_text())
        # Create final list of tags
        for item in items:#tags_list:
            items_list = item.split('(')
            name = items_list[0].rstrip()
            # Get value
            value = item[item.find("(") + 1:item.find(")")]
            dict_of_tags[name] = value
        
    return dict_of_tags
```
